## Construction of the Tangent Plane

We have already discussed how the partial derivative $\dfrac{\partial}{\partial x_{i}}$ is the slope of the **trace**, or slice of a function with respect to $x_{i}$. The graph of $f(x,y)$ is a 3-dimensional **surface** when plotted in 3 dimensions, and hence the construction of all partial derivatives forms a *tangent plane* to the surface.

Letâ€™s start out with a point $(x_0,y_0)$ and let function $C_1$ represent the trace to $f(x,y_0)$ for the plane along $(y_0)$ (i.e. the slice holding y fixed). Let $C_2$ represent the trace to $f(x_0,y)$ for the plane along $(x_0)$ (i.e. the slice with x held fixed).  Now, we know that 

$$f_{x}(x_0,y_0) = \dfrac{\partial f(x_0,y_0)}{\partial x}$$

is the slope of the tangent line to $C_1$ and 

$$f_{y}(x_0,y_0) = \dfrac{\partial f(x_0,y_0)}{\partial y}$$

is the slope of the tangent line to $C_2$. From these slopes we make the tangent lines $L1$ and $L2$, held tangent to $C_1$ and $C_2$. Thus, the plane tangent to the surface at $(x_0,y_0)$ is that plane that contains $L1$ and $L2$. 

### The tangent plane equation

Taking the equation of the plane:

$$0 = a(x-x_0)+b(y-y_0)+c(z-z_0)$$

We can rework it into relationships of slopes relative to z:

$$(z-z_0) = -\frac{a}{c}(x-x_0)-\frac{b}{c}(y-y_0)$$

And generalize them into simple coefficients.

$$(z-z_0) = A(x-x_0)-B(y-y_0)$$

Thus the plane that contains $L1$ and $L2$ simply includes their slope coefficients:

$$z-z_0 = f_{x}(x_0,y_0)(x-x_0)+f_{y}(x_0,y_0)(y-y_0)$$

With $z_0 = f(x_0,y_0)$ naturally,

$$z = f(x_0,y_0)+f_{x}(x_0,y_0)(x-x_0)+f_{y}(x_0,y_0)(y-y_0)$$

## The Gradient Equation and Normal Vector

Just as a plane can be represented by its normal in point-slope form, a gradient produces a normal vector to a surface. 

For a given function $f(x,y)$, the gradient vector $\vec{\nabla} f$ at a given point (x_0,y_0) is normal to the level curve $f(x,y) = k$. Likewise $\vec{\nabla} f(x_0,y_0,z_0)$ is orthogonal to the surface $f(x, y, z) = k$ at $(x_0,y_0,z_0)$

Beginning with:

$$f(x,y) = z$$

$$f(x,y) - z = 0$$

Now we can define a new function:

$$F(x,y) = f(x,y) - z$$

$$F_x(x,y) = \dfrac{\partial}{\partial x}(f(x,y) - z) = f_x$$

$$F_y(x,y) = \dfrac{\partial}{\partial y}(f(x,y) - z) = f_y$$

$$F_z(x,y) = \dfrac{\partial}{\partial z}(f(x,y) - z) = -1$$

The equation for the tangent plane is:

$$f_{x}(x_0,y_0)(x-x_0)+f_{y}(x_0,y_0)(y-y_0)-(z-z_0)=0$$

Solving for $z$ we get:

$$z = f(x_0,y_0)+f_{x}(x_0,y_0)(x-x_0)+f_{y}(x_0,y_0)(y-y_0)$$

Which is the standard equation of the plane.

Thus the equation of the normal line in any field is:

$$\vec{r}(t) = \vec{x_0} + t\vec{\nabla}f$$

## Critical Points

Partial derivatives can be used to find the critical points in a surface in a local neighborhood. Recall that the first derivative provides the slope of the tangent line to a trace. The second derivative provides the curvature of the trace. Partial derivatives can be used in exactly the same way.

### Relative Maxima and Minima

A function $f(x,y)$ has a relative **minimum** at the point $(x_0,y_0)$ if $f(x,y) \geq f(x_0,y_0)$ for all points $(x,y)$ in some region $R$ around $(x_0,y_0)$.

A function $f(x,y)$ has a relative **maximum** at the point $(x_0,y_0)$ if $f(x,y) \leq f(x_0,y_0)$ for all points $(x,y)$ in some region $R$ around $(x_0,y_0)$.

This definition makes no claims about the **global** maximum or minimum value of the function over the entire field. 

The locations of the relative maxima and minima are called **critical points**. A critical point can be found wherever the first derivative is zero. For a function of two or more variables, we can find the equivalent points in the field wherever the gradient is zero:

$$\vec{\nabla}f(x,y) = 0$$

$$\begin{bmatrix}f_x(x,y)\\f_y(x,y)\end{bmatrix} = \begin{bmatrix}0\\0\end{bmatrix}$$


**or** any points (a,b) where $f_{a,b}$ or $f_{a,b}$ don't exist.

Note that all of the first partial derivatives must be 0 at (a, b)

Therefore, if $(a,b)$ is a relative extrema of the function $f(x,y)$ then $(a,b)$ is also a critical point of $f$. And so:

$$\vec{\nabla}f(a,b) = 0$$

### Saddle Points

There are many functions that have critical points at a point, say $(a,b)$ (first derivative 0 in all coordinates), but have quadrants where there are points on the surface with values larger than $f(a,b)$ and smaller than $f(a,b)$. Therefore, this critical point is not a relative extremum. It is called a **saddle point**. 

How can we tell if a critical point is a saddle point? We use the *curvature* or second derivative. Recall from calculus that the second derivative describes the curvature of the trace. 

The curvature of a saddle point of a function in three coordinates has to be determined in relationship to both the independent coordinates.

When $(a,b)$ is a critical point of $f(x,y)$ and the second order partial derivatives are continuous in some region that contains $(a,b)$. Next we define a *measure* $D$:

$$D(a,b) = f_{xx}(a,b)f_{yy}(a,b) - [f_{xy}(a,b)]^{2}$$

Using $D$ and the second-order partial derivatives, we can find and classify the saddle points and relative extrema at any point on the surface. 

For a given point on the surface $f(x,y)$

1. $D>0$ and $f_{xx}(a,b) > 0$ or  $f_{yy}(a,b) > 0$, then $(a,b)$ is a relative **minimum**.
2. $D>0$ and $f_{xx}(a,b) < 0$ or  $f_{yy}(a,b) < 0$, then $(a,b)$ is a relative **maximum**.
3. $D<0$ then $(a,b)$ is a saddle point
4. $D=0$ then we cannot classify the point directly. We would need to check nearby points and nearby trace curvature to determine its behavior.

## Example:

Find and classify all critical points of $f(x,y) = 4+x^3+y^3-3xy$.

$f_{x} = 3x^2-3y$ (first partial derivate respect x)

$f_{y} = 3y^2-3x$ (first partial derivate respect y)

$f_{xx} = 6x$ (second partial derivate respect x)

$f_{yy} = 6y$ (second partial derivate respect y)

$f_{xy} = -3$ (second partial derivate of x respect y)

Critical points will be all the solutions to the system of equations (first partial derivative equations for x and y):

$3x^2-3y = 0$

$3y^2-3x = 0$

we can see:

$y = x^2$

Plugging into the second equation:

$3(x^2)^2-3x  = 0$

$ 3x(x^3-1) = 0$

$x=0$ or $x=1$, since $y=x^2$, we have two critical points:

$(1,1)$ or $(0,0)$

for both points we calculate $D$:

$$D(x,y) = (6x)(6y)-(-3)^2$$

$$D(x,y) = 36xy-9$$

$D(0,0) = -9$ is negative and so this is a saddle point.

$D(1,1) = 25$ is positive and $f_{xx}(1,1) = 6$ so this is a relative minimum.
