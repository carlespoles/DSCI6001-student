##Equations of lines

We will be discussing parametric equations several times during this term. Parametric equations describe the *pathway* of an otherwise normal linear equation in terms of a single variable, or *parameter*. The parameter (often given as $t$) is something like the coefficients belonging to a span, in that it is assumed to encompass the whole variable space to which it belongs (i.e. all of $R$).

There are several types of parametric equations:

1. Equations of straight lines in $R^{N}$
2. Equations of curvilinear lines in $R^{N}$
3. Equations of systems
4. Equations of structures

We will only cover the first three. There are many more. 

Why do we use parametric equations? Because the use of parameters simplifies the analysis of the system. It frequently occurs that derivatives are easier to take and understand when related to a single variable (the parameter) rather than many variables.

The first type of equation, that of a straight line, can be understood in terms of the below figure:

![parametric-1d](./images/parametric_line_2d.png)

We can write the equation of the line as part of a process controlled by the parameter $\vec{d}$:

$$\begin{bmatrix}x-x_{0}\\ y-y_{0}\end{bmatrix} = t\begin{bmatrix}d_{x}\\d_{y}\end{bmatrix}$$

That vector $\vec{d}$ is the unit vector in the direction of the line. The parameter simply multiplies it by some coefficient that we choose. We can break this expression up into proper vector sums using the unit vector:

$$x-x_{0} = td_{x}$$
$$y-y_{0} = td_{y}$$

$$\dfrac{x-x_{0}}{d_{x}} = \dfrac{y-y_{0}}{d_{y}}$$

###Normal Equations of Lines

Lines and planes are directly related. A second way to characterize a line is to give a point on the line and a normal vector (ostensibly emerging from the point).

What is the equation of the normal to the line vector? Well one way of determining it is to compute it from the dot product, thus we define the normal as any vector which is perpendicular to the line:

$$\vec{v} = \begin{bmatrix}x-x_{0}\\y-y_{0}\end{bmatrix}$$

$$n_{\vec{v}} = \begin{bmatrix}n_{x}\\n_{y}\end{bmatrix}$$

$$n_{\vec{v}} \cdot \vec{v} = 0$$

$$n_{x}(x-x_{0})+n_{y}(y-y_{0}) = 0$$

$$n_{x}x+n_{y}y=  n_{x}x_{0}+n_{y}y_{0}$$

##Example 1:

Suppose we have a line with the equation $-4x + 3 = y$. Produce parametric and normal versions of this line. 

First, generate the equation of the line from its vector. Generate two points on the line and take their difference to find the vector:A point on this line is $(2, -5)$. 

$(2,-5)$ and $(4, -13)$ are both points on the line. The vector describing the line is described as: 

$\vec{v} = \begin{bmatrix}4-2\\-13-(-5)\end{bmatrix} = \begin{bmatrix}2\\-8\end{bmatrix}$

So the parametric equation is written as (please note that we choose $x_0 = 0$ and $y_0 = 3$. When x = 0 in $-4x + 3 = y$, y = 3):

$t\begin{bmatrix}1\\-4\end{bmatrix}+\begin{bmatrix}0\\3\end{bmatrix} = \begin{bmatrix}x\\y\end{bmatrix}$

Rewritting above equation:

$-4x + 3 = y, -4x - y = -3, 4x + y = 3$, so normal vector is made out of the coefficients of x (4) and y (1):

$[4,1]^T$

One way to determine a normal to the line is simply to intuit one. Any vector whose dot product with the line's vector was zero would be eligible. Therefore,

$$\vec{n} = \begin{bmatrix}4\\1\end{bmatrix}$$ 

is one candidate. 

We can generate another candidate normal vector by using the method given in the handout, which is to set everything to look like the above equation:

$$\begin{bmatrix}-4&-1\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix} = -3$$

The point $(0,3)$ satisfies the above linear equation. To double check we can plug this in for $(x_{0},y_{0})$.

To make it look like our other normal, we just multiply the above equations through by -1:

$$\begin{bmatrix}4&1\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix} = 3$$

$$\begin{bmatrix}4&1\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix} = \begin{bmatrix}4&1\end{bmatrix}\begin{bmatrix}0\\3\end{bmatrix}$$

$$\begin{bmatrix}4&1\end{bmatrix}\begin{bmatrix}x-0\\y-3\end{bmatrix} = 0$$

While it may seem weird at first to have a bunch of properly opposing vectors pop out of the same equation, this is indeed what is supposed to happen.

## Equations of Planes

We can simply extend the logic and notation we use for lines directly to 3 dimensional planes by adding in a third coordinate:

$$n_{x}x+n_{y}y+n_{z}z =  n_{x}x_{0}+n_{y}y_{0}+n_{z}z_{0}$$

(a line is a plane in two dimensions).

This formula is precisely the same as the one we learn in highschool,

$$ax+by+cz = d$$


Unlike a line however, simply finding a vector that lies along the plane and a point on the plane does not uniquely determine the plane, because it can rotate freely about the axis of the vector. Only the normal vector to the plane and a point on the plane will do the trick.

## Explanation of describing a plane with 2 points and a normal to the points
2 points in a plane: (x,y,z) and ($x_0,y_0,z_0$)

The vector normal to the points is [$n_x,n_y,n_z$]

The dot product of the vector defined by the points and the normal vector is 0:

[$x-x_0,y-y_0,z-z_0] * [n_x,n_y,n_z] = 0, n_x * x + n_y * y + n_z * z = x_0 * n_x + y_o * n_y + z_0 * n_z$

###Example 1:

Given that (-8, 4) and (4, 3) lie on the same line, what is the equation of the line?

(1) First produce the directional vector pointing between the two points.

$$\vec{d} = \begin{bmatrix}12 \\ -1\end{bmatrix}\ or\ \begin{bmatrix}-12 \\ 1\end{bmatrix} $$

(2) Second, determine the intercept. The slope of the line is -1/12 (which is $\frac{d_y}{d_x} = -1/12$), thus we can write the equation as:
    
$$(y-y_{1}) = -\frac{1}{12}(x-x_{1})$$
    
Using one of the above points (I like (4,3)), we can write:
    
$$(y-3) = -\frac{1}{12}(x-4)$$

Simplifying we get

$$12y + x = 40\ or\ y = \frac{1}{12}(40-x)$$


Another way of doing above:

$\begin{bmatrix}x-4 \\ y-3\end{bmatrix} = t * \begin{bmatrix}12 \\ -1\end{bmatrix}$

then x-4 = t(12) and y-3 = t(-1),

t = 3-y

###Example 2:

We are given the 2-dimensional parametric lines $\begin{bmatrix}x\\y\end{bmatrix} = t\begin{bmatrix}7\\2\end{bmatrix} + \begin{bmatrix}5\\-1\end{bmatrix}$ and $\begin{bmatrix}x\\y\end{bmatrix} = t\begin{bmatrix}3\\9\end{bmatrix} + \begin{bmatrix}-3\\0\end{bmatrix}$. 

1. Do these intersect in $R^2$?
2. If so, where do they intersect?

(1) First, determine if they intersect. This can always be done by determining if their simultaneous equations have a solution.
(**QUIZ** How do you do that?) 
In $R^2$, we are assured that two lines intersect if they do not have the same slope because this assures us that they are not parallel. In general, in any coordinate system, all we have to know is if the two lines are exactly parallel or not to determine if they intersect(why?).

The slope of the first line is $\frac{2}{7}$ and the second line is $3 (= \frac{9}{3})$


(2) To determine where the lines intersect we need to find a solution for both lines simultaneously....

First we recompose the parametric equations as linear equations and factor out the parameter $t$:

$$x-x_{0} = td_{x}$$
$$y-y_{0} = td_{y}$$

$$\dfrac{x-x_{0}}{d_{x}} = \dfrac{y-y_{0}}{d_{y}}$$

$$(x-x_{0})d_{y} - (y-y_{0})d_{x} = 0$$

Then transforming to the linear equation:

$$n_{x}(x-x_{0}) + n_{y}(y-y_{0}) = 0$$

$$n_{x}x+ n_{y}y = n_{x}x_{0}+n_{y}y_{0} $$

Here we get for $\begin{bmatrix}x\\y\end{bmatrix} = t\begin{bmatrix}7\\2\end{bmatrix} + \begin{bmatrix}5\\-1\end{bmatrix}$:

$$2(x-5)-7(y+1) = 0$$

$$7y = 2x-17$$

For $\begin{bmatrix}x\\y\end{bmatrix} = t\begin{bmatrix}3\\9\end{bmatrix} + \begin{bmatrix}-3\\0\end{bmatrix}$:

$$9(x+3)-3(y) = 0$$

$$y = 3x + 9$$

You need to find the place where $y_{eq1} = y_{eq2}$ and $x_{eq1} = x_{eq2}$

In this case we can set **simultaneous equations**:

$$\begin{bmatrix}-2&7\\-3&1\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix}=\begin{bmatrix}-17\\9\end{bmatrix}$$

The **determinant of the matrix is nonzero** so we can compose the inverse and solve the linear equation $A\vec{x} = b$ to give $\vec{x} = A^{-1}b$.

Using the generalized inverse, we get:

$$A^{-1} = \dfrac{1}{19}\begin{bmatrix}1&-7\\3&-2\end{bmatrix}$$

Thus:

$$\begin{bmatrix}x\\y\end{bmatrix}=\dfrac{1}{19}\begin{bmatrix}1&-7\\3&-2\end{bmatrix}\begin{bmatrix}-17\\9\end{bmatrix} = \dfrac{1}{19}\begin{bmatrix}-80\\-69\end{bmatrix}$$



###Example 3:

We are given two directional vectors that span a plane, $\vec{v} = \begin{bmatrix}1 \\ -1 \\ 1\end{bmatrix}$, $\vec{w} = \begin{bmatrix}1 \\ 3 \\ 2\end{bmatrix}$  

And a point on the plane, $(-4,3,2)$

What is the equation for the plane?

(1) First you need the normal to the plane, which can be gotten through the cross product of the two vectors:

$$\vec{v} \times \vec{w} = \begin{bmatrix}v_{2}w_{3}-v_{3}w_{2}\\v_{3}w_{1}-v_{1}w_{3}\\v_{1}w_{2}-v_{2}w_{1}\end{bmatrix}$$

$$\vec{v} \times \vec{w} = \begin{bmatrix}-5\\-1\\4\end{bmatrix}$$

(2) A point contained on this plane must be found within the plane as an origin, therefore: $-5(x-(-4))-(y-3)+4(z-2) = 0$, is the equation of the plane: $-5x-y+4z=25$.

####QUIZ:
Prove that this is the correct equation for the plane by determining that the equation produces lines that are in the plane of the spanning vectors.


**Answer:**
One point that satisfies the calculated equation is $(-1,0,5)$. We can make a vector using this point and given point, $(-4,3,2)$, 

$$\vec{k} = \begin{bmatrix}-3\\3\\-3\end{bmatrix}$$

If $\vec{k}$ is perpendicular to the normal of the plane, then it lies in the plane. $\vec{k} \cdot (\vec{v} \times \vec{w}) = 0$, so this is true.

###Example 5:

We are given a point $P$: $(6,-4,1)$ and a plane $U$: $3x+5y-9z-1=0$. What is the (shortest) distance between this line and the surface of the plane?

This can be an intimidating problem, but it uses the same logic that we have applied before. 

(1) First, we check that the point does not lie in the plane. $18-20-9 \neq 1$, so we can be sure it is not in the plane.

(2) Second, since the plane is infinite, we can choose the relationship between the point and the normal of the plane as the metric of measurement (since the normal is the same everywhere). We need to measure the distance from our target point to any point on the plane (it doesn't matter where - why?)

The normal of the plane is $\vec{n_{U}} = \begin{bmatrix}3\\5\\-9\end{bmatrix}$ and an arbitrary point on the plane (pick something easy) is $K$: $(0,2,1)$. Now take the projection of the vector from $P$ pointing to $K$:

$$\vec{z} = \begin{bmatrix}6\\-6\\0\end{bmatrix} = \begin{bmatrix}6-0\\-4-2\\1-1\end{bmatrix}$$

onto $n_{U}$,

$$proj_{\vec{n_{U}}}\vec{z} = \dfrac{\vec{z} \cdot \vec{n_{U}} }{\|\vec{n_{U}}\|}$$

All we need is the magnitude of this projection:

$$|proj_{\vec{n_{U}}}\vec{z}| = \dfrac{|3\cdot6+5\cdot(-6)-9\cdot0|}{\sqrt{9+25+81}}$$

$$|proj_{\vec{n_{U}}}\vec{z}| = \dfrac{12}{\sqrt{115}}$$

In general, given a point $(x_{0},y_{0},z_{0})$ and the equation of a plane, $ax+by+cz+d = 0$, we can calculate the distance $D$ from the point to the plane as:

$$D = \dfrac{ax_{0}+by_{0}+cz_{0}+d}{\sqrt{a^2+b^2+c^2}}$$

We can use this equation here for our amusement:

$$D = \dfrac{3\cdot6+5\cdot(-4)-9\cdot1-1}{\sqrt{3^2+5^2+(-9)^2}} = \dfrac{3\cdot6+5\cdot(-4)-9\cdot1-1}{\sqrt{3^2+5^2+(-9)^2}}=\dfrac{12}{\sqrt{115}}$$
