## Eigendecomposition or Spectral Decomposition

In the last lecture we discussed the diagonalization of a matrix:

$${\bf A} = {\bf PDP}^{-1}$$

To be consistent with the standard notation in the context of diagonalization, we'll represent the matrix of eigenvectors by P:

$${\bf P} = \left[ \begin{matrix} {\bf u}_1 & \cdots & {\bf u}_n \end{matrix} \right]$$

As before, ${\bf u}_1, \cdots, {\bf u}_n$ are eigenvectors of $\bf A$ that form a basis for $R^n$. to convert a vector from standard basis to basis $B$, we compute

$${\bf x}_B = {\bf P}^{-1} {\bf x}$$

Then

$${\bf D x}_B = {\bf D P}^{-1}{\bf x}$$

Finally, to convert back to the standard basis, we multiply by $\bf P$:

$${\bf Ax} = {\bf PDP}^{-1} {\bf x}$$

This holds for all vectors $\bf x$, therefore $\bf A = {\bf PDP}^{-1}$

**Example:**

Consider a $2\times 2$ matrix:

$${\bf A} = \left[ \begin{matrix} 3 & 1 \\ -2 & 0 \end{matrix} \right]$$

To diagonalize $\bf A$, we first compute the eigenvalues and eigenvectors:

$$\text{det}({\bf A} - \lambda {\bf I}) = \left| \begin{matrix} 3 - \lambda & 1 \\ -2 & -\lambda \end{matrix} \right | = \lambda^2 - 3\lambda +2$$

Setting the determinant to zero, we get two eigenvalues $\lambda_1 = 1$ and $\lambda_2 = 2$. Starting with $\lambda_1 = 1$. The eigenvectors corresponding to these eigenvalues are

$${\bf u}_1 = \left[ \begin{matrix} 1 \\ -2 \end{matrix} \right], ~~~~\text{and} ~~~~ {\bf u}_2 = \left[ \begin{matrix} 1 \\ -1 \end{matrix} \right]$$

Now we define $\bf P$ and $\bf D$, with

$${\bf P} = \left[ \begin{matrix} {\bf u}_1 & {\bf u}_2 \end{matrix} \right] = \left[ \begin{matrix} 1 & 1 \\ -2 & -1 \end{matrix} \right]~~~~\text{and}~~~~{\bf D} = \left[ \begin{matrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{matrix} \right] = \left[ \begin{matrix} 1 & 0 \\ 0 & 2 \end{matrix} \right]$$

Check:

$${\bf PDP}^{-1} =  \left[ \begin{matrix} 1 & 1 \\ -2 & -1 \end{matrix} \right] \left[ \begin{matrix} 1 & 0 \\ 0 & 2 \end{matrix} \right] \left[ \begin{matrix} -1 & -1 \\ 2 & 1 \end{matrix} \right]  = \left[ \begin{matrix} 3 & 1 \\ -2 & 0 \end{matrix} \right] = {\bf A}$$

## Construction of a matrix with specified Eigenvectors

Another illustration of diagonalization is the fact that the diagonalization formula can also be used to construct a matrix that has specified eigenvalues and eigenvectors. 

**Example:**

Let $\lambda_1 = -1$ and $\lambda_2 = 2$ be the eigenvalues corresponding to the eigenvectors ${\bf u}_1 = \begin{bmatrix}5\\3\end{bmatrix}$ and ${\bf u}_2 = \begin{bmatrix}3\\2\end{bmatrix}$. We wish to find a matrix $\bf A$ which gives these eigenvalues and eigenvectors.

We start by constructing $\bf P$ and $\bf D$:

$${\bf P} = \left[ \begin{matrix} {\bf u}_1 & {\bf u}_2 \end{matrix} \right] = \left[ \begin{matrix} 5 & 3 \\ 3 & 2 \end{matrix} \right]~~~~\text{and}~~~~{\bf D} = \left[ \begin{matrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{matrix} \right] = \left[ \begin{matrix} -1 & 0 \\ 0 & 2 \end{matrix} \right]$$

Thus

$${\bf P}^{-1} = \left[ \begin{matrix} 2 & -3 \\ -3 & 5 \end{matrix} \right]$$

This gives

$${\bf A} = {\bf PDP}^{-1} = \left[ \begin{matrix} 5 & 3 \\ 3 & 2 \end{matrix} \right]\left[ \begin{matrix} \lambda_1 & 0 \\ 0 & \lambda_2 \end{matrix} \right]\left[ \begin{matrix} 2 & -3 \\ -3 & 5 \end{matrix} \right] = \left[ \begin{matrix} -28 & 45 \\ -18 & 29 \end{matrix} \right]$$

Check:

$${\bf A u}_1 = \left[ \begin{matrix} -28 & 45 \\ -18 & 29 \end{matrix} \right] \left[ \begin{matrix} 5 \\ 3 \end{matrix} \right] = \left[ \begin{matrix} -5 \\ -3 \end{matrix} \right] = (-1)\left[ \begin{matrix} 5 \\ 3 \end{matrix} \right]$$

$${\bf A u}_2 = \left[ \begin{matrix} -28 & 45 \\ -18 & 29 \end{matrix} \right] \left[ \begin{matrix} 3 \\ 2 \end{matrix} \right] = \left[ \begin{matrix} 6 \\ 4 \end{matrix} \right] = (2)\left[ \begin{matrix} 3 \\ 2 \end{matrix} \right]$$

Hence $\lambda_1$ and $\lambda_2$ are eigenvalues of $\bf A$ associated with eigenvectors ${\bf u}_1$ and ${\bf u}_2$.

## Diagonalization and Distinct Eigenvalues

If a matrix $\bf A$ has no repeated eigenvalues, it has $n$ distinct eigenvalues.

In this case the $n$ eigenvectors are automatically linearly independent and form a full basis for $R^n$. **Therefore, any matrix with distinct eigenvalues can be diagonalized.**

Not all matrices possess $n$ linearly independent eigenvectors, thus **not all matrices are diagonalizable, including invertible matrices.** 

For example, consider the following nondiagonalizable matrix:

$${\bf A} = \left[ \begin{matrix} 0 & 1 \\ 0 & 0 \end{matrix} \right]$$

Its eigenvalues are $\lambda_1 = \lambda_2 = 0$, and there is only one independent eigenvector, $\begin{bmatrix}1\\0\end{bmatrix}$. 

## Repeated eigenvalues do not always lead to nondiagonalizability

Consider the following $3\times 3$ matrix:

$${\bf A} = \left[ \begin{matrix} 1 & 1 & 1 \\ -2 & -2 & -1 \\ 0 & 0 & -1 \end{matrix} \right]$$

The characteristic polynomial of $\bf A$ is

$$\text{det}({\bf A} - \lambda {\bf I}) = -\lambda^3 -2\lambda^2 -1 = -\lambda(\lambda +1)^2$$

Thus we have eigenvalues $\lambda_1 = 0$, and $\lambda_2 = \lambda_3 = -1$.

The eigenvector corresponding to eigenvalue 0 is

$${\bf u}_1 = \left[ \begin{matrix} 1 \\ -1 \\ 0 \end{matrix} \right]$$

For eigenvalue -1, we can find two independent eigenvectors:

$${\bf u}_2 = \left[ \begin{matrix} 1 \\ -2  \\ 0 \end{matrix} \right], ~~~~\text{and} ~~~~ {\bf u}_3 = \left[ \begin{matrix} 1 \\ 0  \\ -2 \end{matrix} \right]$$

Hence the set $\{{\bf u}_1,{\bf u}_2,{\bf u}_3\}$ is linearly independent and thus form a basis for $R^3$. From these eignevalues and eigenvectors we can construct $\bf P$ and $\bf D$ as

$${\bf P} = \left[ \begin{matrix} 1 & 1 & 1 \\ -1 & -2 & 0 \\ 0 & 0 & -2 \end{matrix} \right], ~~~~\text{and}~~~~{\bf D} = \left[ \begin{matrix} 0 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & -1 \end{matrix} \right]$$

Since there are two eigenvectors corresponding to eigenvalue $-1$, this eigenvalue apprears twice in $\bf D$.

### Matrix Powers

Suppose that $\bf A$ is diagonalizable. Then

$${\bf A}^2 = ({\bf PDP}^{-1})({\bf PDP}^{-1}) = {\bf PD} ({\bf P}^{-1} {\bf P}){\bf DP}^{-1} = {\bf PD}^2 {\bf P}^{-1}$$

Similarly

$${\bf A}^3 = {\bf A A}^2 = ({\bf PDP}^{-1})({\bf PD}^2 {\bf P}^{-1}) = {\bf PD}^3 {\bf P}^{-1}$$

In general

$${\bf A}^k = {\bf PD}^k {\bf P}^{-1}$$

where ${\bf D}^k$ is simply

$${\bf D} = \left[ \begin{matrix} d_{11} & 0 & \cdots & 0 \\ 0 & d_{22} & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & d_{nn} \end{matrix} \right]~~~~~\implies~~~~ {\bf D}^k = \left[ \begin{matrix} d^k_{11} & 0 & \cdots & 0 \\ 0 & d^k_{22} & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & d^k_{nn} \end{matrix} \right]$$

This is a powerful conclusion. This means that $n$ applications of the matrix as a transformation is the same as applying $n$ powers of the deformation matrix, with the same "sandwich" of the basis-rotation matrices.
