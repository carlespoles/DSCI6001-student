## Composite Functions

The theory of functions is a really deep rabbithole and we cannot spend too much time on them. Many functions can be *de-composed* or **decomposed** as a composite, or grouping, of several other more *elementary* [functions]. This is a very helpful way to think of calculus in general. 

We say a given function $h(x)$ is equal to $f(x)$ composed of $g(x)$:

$$h(x) = f \circ g = f(g(x))$$

This is to say simply that the inputs of $f(x)$ are the outputs of $g(x)$. Obviously with non-bijective functions, a number of functions with relatively crazy domains and ranges are possible. 

### Example:

$g(x) = 4x^2$

$f(x) = e^x$

$f \circ g = e^{4x^2}$

## Chain Rule

Derivatives can help us get around the issue of domains and ranges a bit.
Suppose $f(x) = \sqrt{x^2 + 1}$ and $g(x) = 2x + 1$, and we have

$$y = f(g(x)) = \sqrt{(2x+1)^2 + 1}$$

If $f$ and $g$ are differentiable, then so is the composite function. The derivative of the composit function can be obtained using _chain rule_:

If $y=f(u)$ is a differentiable function of $u$ and $u=g(x)$ is in turn a differentiable function of  $x$, then the composit function $y=f(g(x))$ is a differentiable function of $x$ whose derivative is given by the product

$$\frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx}$$


or, equivalently, by

$$\frac{dy}{dx} = f'(g(x)) g'(x)$$


#### Example

We want to find the derivative of the function:

$$y=f(x) = (x^2+2)^2 - 3 (x^2+2) + 1$$

Note that $y = u^2 - 3 u + 1$, where $u = x^2 + 2$. Thus

$$ \frac{dy}{du} = 2 u - 3 ~~~~~~~~~ \text{and}~~~~~~~~~~ \frac{du}{dx} = 2x$$

and according to the chain rule:

$$\frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx} = (2u-3)(2x) = [2(x^2+2) - 3](2x) = 2x(2x^2+1)$$

#### Conceptual Insight

The key aspect of the chain rule is that *rates of change multiply.* To illustrate, suppose that your salary increases twice as fast as your friend's. If your friend's salary increases \$4000 per year, your salary will increase at the rate of 2$\times$4000 or \$8000 per year. In terms of derivatives,

$$\frac{d(\text{your salary})}{dt} = \frac{d(\text{your salary})}{d(\text{friend's salary})} \times \frac{d(\text{friend's salary})}{dt}$$

$$ $8000/\text{yr} = 2 \times $4000/\text{yr}$$

As another example, imagine a circle whose radius increases at a rate of $3~cm/s$. We can find the rate at which the area is increasing using the chain rule:

$$\frac{dA}{dt} = \frac{dA}{dr} \frac{dr}{dt}$$

We're given the rate $dr/dt$, namely $dr/dt = 3~cm/s$. To compute $dA/dr$, we use the formula for the area of the circle, $A = \pi r^2$.

$$\frac{dA}{dr} = \frac{d}{dr}(\pi r^2) = 2\pi r$$


Because, $dr/dt = 3~cm/s$, we obtain

$$\frac{dA}{dt} = (2\pi r)(3) = 6\pi r ~ cm^2/s$$


### QUIZ:

The cost of producing $x$ units of a particular commodity is $C(x) = \frac{1}{3}x^2 + 4x + 53$ dollars, and the production level $t$ hours into a particular produciton run is $x(t) = 0.2 t^2 + 0.03 t$ units. At what rate is cost changing with respect to time after 4 hours?

We find that

$$\frac{dC}{dx} = \dfrac{2}{3}x + 4$$

and 

$$\frac{dx}{dt} = 0.4 t + 0.03$$

so according to the chain rule

$$\frac{dC}{dt} = \frac{dC}{dx} \frac{dx}{dt} = \left( \frac{2}{3}x + 4\right) (0.4 t + 0.03)$$

When $t=4$, the level of production is

$$x(4) = 0.2(4)^2 + 0.03(4) = 3.32~\text{units}$$

and by substituting $t=4$ and $x=3.32$ into the formula for $dC/dt$, we get

$$\left. \frac{dC}{dt}\right|_{t=4} = \left[ \frac{2}{3} (3.32) + 4 \right][0.4(4) + 0.03] = 10.77$$

## Chain Rule for Composite Functions

If $z=f(x,y)$ is a function of variables $x$ and $y$ with partial derivativs $f_x$ and $f_y$, and if $x$ and $y$ are differentiable functions of another varialbe $t$, i.e., $x=g(t)$ and $y=h(t)$, then $z$ is also a function of $t$. In this case, we can find the _total derivative_ of $z$ w.r.t to $t$ given by

$$\frac{dz}{dt} = \frac{\partial z}{\partial x} \frac{dx}{dt} + \frac{\partial z}{\partial y} \frac{dy}{dt}$$

Note that the expression for $dz/dt$ is the sum of two terms, each of which can ï¿¼be interpreted using the chain rule for a function of one variable.

$$\frac{\partial z}{\partial x} \frac{dx}{dt} = \text{rate of change of $z$ with respect to $t$ for fixed $y$}$$

$$\frac{\partial z}{\partial y} \frac{dy}{dt} = \text{rate of change of $z$ with respect to $t$ for fixed $x$}$$

The chain rule for partial derivatives says that the total rate of change of $z$ with respect to $t$ is the sum of these two _partial_ rates of change.

The above definition can be extended to functions of more than two variables. If $w=f(x,y,z,...)$ is a function of several varialbes, $x,y,z,...$, which in turn are functions of variable $t$, then the total derivative of $w$ is

$$\frac{dw}{dt} = \frac{\partial w}{\partial x} \frac{dx}{dt} + \frac{\partial w}{\partial y} \frac{dy}{dt} + \frac{\partial w}{\partial z} \frac{dz}{dt} + \cdots$$


#### Example 1

Find $dz/dt$, given $z=x^2 + 3 x y + 2 y^2$; $x = t^2$, $y = 2t+1$.

Solution:

$$\frac{\partial z}{\partial x} = 2 x + 3,~~~~ \frac{\partial z}{\partial y} = 3x + 4 y$$

$$\frac{dx}{dt} = 2 t, ~~~~ \frac{dy}{dt} = 2$$


Using
$$\frac{dz}{dt} = \frac{\partial z}{\partial x} \frac{dx}{dt} + \frac{\partial z}{\partial y} \frac{dy}{dt}$$

We get,
$$\frac{dz}{dt} = (2 x + 3)(2t) + (3x + 4y)(2) = (2 t^2 + 3)(2t) + (3 t^2 + 4(2t+1))(2)$$

$$\frac{dz}{dt} = 2(2 t^3 + 3 t^2 + 7 t + 4)$$

Note that the total derivative is a function of $t$.


#### Example 2

Find (a) $\frac{dz}{dx}$ and (b) $\frac{dz}{dy}$, given $z=f(x,y) = x y^2 + x^2 y$, $y = \ln x$.

(a) Here $x$ is the independent variable. So we'll modify the chain rule as:

$$\frac{dz}{dx} = \frac{\partial z}{\partial x} + \frac{\partial z}{\partial y}\frac{dy}{dx}$$

$$\frac{\partial z}{\partial x} = y^2 + 2 x y, \frac{\partial z}{\partial y} = 2 x y + x^2, \frac{dy}{dx} = \frac{1}{x}$$

Putting it all together, we get:

$$\frac{d z}{d x} = (y^2 + 2 x y) + (2 x y + x^2)\frac{1}{x} = y^2 + 2xy + 2y +x$$

(b) Here $y$ is the independent variable:

$$\frac{dz}{dy} = \frac{\partial z}{\partial x}\frac{dx}{dy} + \frac{\partial z}{\partial y}$$

Since $y = \ln x$, this implise $x = e^y$. Therefore,

$$\frac{dx}{dy} = \frac{d}{dy}(e^y) = e^y = x$$

Thus

$$\frac{dz}{dy} = (y^2 + 2 x y) x + (2 x y + x^2) = x y^2 + 2 x^2 y + 2 x y + x^2$$
