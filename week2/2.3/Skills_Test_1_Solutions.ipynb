## Skills Test 1 Solutions

### 1.


Show that if $ad-bc = 0$ then the below system of equations:

$$ax_1 + bx_2 = r$$
$$cx_1 + dx_2 = s$$

Is consistent. You may make reasonable assumptions about $a, d, b$ and $c$.

(solution)

If $ad - bc = 0$, then ad=bc. If this is true then we may make simplifying assumptions about these variables. 

$a = bc/d$, meaning that d cannot be 0 if s is not 0

Now performing row reduction, 

$$ax_1 + bx_2 = r$$
$$0 + d-\dfrac{cb}{a}x_2 = s-\dfrac{cb}{a}$$

This set of reductions takes on several possible solution states.

in general

$$ \text{if} \ r, s \neq 0; \text{then} \dfrac{a}{c} = \dfrac{b}{d}$$

We can say more about this system, however, if we want to do a little more row reduction. Set $ a = \dfrac{bc}{d}$, as above, then substitute (written in matrix form)!

$$
  \left[\begin{array}{rr|r}
             \dfrac{bc}{d} & b & r \\
                  c & d & s 
  \end{array}\right]
$$

Doing row reduction gets us the following:

$$
  \left[\begin{array}{rr|r}
             \dfrac{bc}{d} & b & r \\
                  0 & 0 & s - \dfrac{dr}{b} 
  \end{array}\right]
$$

Hence the solution space requires that $s = \dfrac{dr}{b}$ 

Which is one working case. The other working case follows when the bottom row is the nonzero row. We can substitute $ d = \dfrac{bc}{a}$ for example. In this case the matrix reduces to:

$$
  \left[\begin{array}{rr|r}
                  0 & 0 & r- \dfrac{sa}{c}\\
                  c & \dfrac{bc}{a} & s 
  \end{array}\right]
$$

In this case the solution space requires that $ r = \dfrac{sa}{c}$ 

( counter case )

Suppose $ad \neq bc$...

We can choose that either $a$ and $d$ or $b$ and $c$ must not be equal to zero. Here, we'll assume that $a \neq 0$, and we'll place the system into an augmented matrix for simplicity:

$\begin{bmatrix}
    a & b & r \\
    c & d & s
\end{bmatrix}$

Now we'll solve to eliminate $c$:

$\begin{bmatrix}
    a & b & r \\
    0 & d-\frac{c}{a}b & s-\frac{c}{a}r \\
\end{bmatrix}$

We know that this system must be consistent because $\dfrac{da-bc}{a} \neq 0$ if $a \neq 0$.

We also need to examine what happens if $a = 0$. In this case it must be true that $b \neq 0$. When this happens, the system looks a little different:

$\begin{bmatrix}
    0 & b & r \\
    c & d & s
\end{bmatrix}$

Here $x_2 = r/b$, giving $x_1 = \dfrac{s-\frac{dr}{b}}{c}$. In this case the system is consistent, assuming $c \neq 0$.

### 2.

$$3x_1-x_2+x_3+2x_4 = -2$$
$$x_1 + 2x_2 - x_3 + x_4 = 1$$
$$-x_1 - 3x_2 + 2x_3 - 4x_4 = -6$$

(solution)

This system is underdetermined but has a solution set. The RREF of the augmented matrix is:

$$\begin{bmatrix}
    1 & 0 & 0 & \frac{5}{3} & 1\\
    0 & 1 & 0 & \frac{-11}{3} & -5 \\
    0 & 0 & 1 & \frac{-20}{3} & -10
\end{bmatrix}$$

This means that the general solution can be held with $x_4$ as the free variable, i.e.

$$\begin{bmatrix}
    1 & 0 & 0 & \frac{5}{3} & 1\\
    0 & 1 & 0 & \frac{-11}{3} & -5 \\
    0 & 0 & 1 & \frac{-20}{3} & -10 \\
    0 & 0 & 0 & 1 & 0
\end{bmatrix}$$

and we can compose the solution as a vector of $x_4 \in \mathbb{R}$ and $b$:

$$\begin{bmatrix}
     x_1\\
     x_2 \\
     x_3 \\
     x_4
\end{bmatrix} = \begin{bmatrix}
     1\\
    -5 \\

\end{bmatrix} + x_4\begin{bmatrix}
     \frac{-5}{3}\\
    \frac{11}{3} \\

\end{bmatrix} $$

This means that an example solution of the infinite possible solutions might be:

\end{bmatrix} = \begin{bmatrix}

 If you dot this vector with the coefficient matrix:
 
 $$\begin{bmatrix}

    \frac{2}{3}+4-\frac{20}{3}-4\\   
 \end{bmatrix} = \begin{bmatrix}

### 4.

Suppose there are two matrices $A$ and $B$ and we may define a relationship between $A$ and $B$ as follows:

$$A = U^{-1}BU$$

This is to say that $A$ and $B$ are *similar matrices*. Show that $det(A) = det(B)$.



(solution)

$$det(A) = det(U^{-1}BU)$$

by the properties of the determinant

$$det(A) = det(U^{-1})det(B)det(U)$$

And so on. 

from here there are several choices that can be made that make similar claims. One can first solve for $BU$, as $U$ is clearly invertible.

$$UA = UU^{-1}BU$$

$$UA = BU$$

$$det(U)det(A) = det(B)det(U)$$

applying commutativity:

$$det(U)det(A) = det(U)det(B)$$

confirming that 

$$det(A) = det(B)$$

Similar treatments can be applied to either side. 

Another choice can collect the determinants of $U$ and its inverse to one side:

$$det(A) = det(U^{-1})det(B)det(U)$$

$$det(A) = det(B)det(U^{-1})det(U)$$

$$det(A) = det(B)det(U^{-1}U)$$

$$det(A) = det(B)$$

This one is more elegant but skips some obvious statements about the matrix product.

### 5. 

A $n \times n$ *square* matrix is **skew-symmetric** if $A^{T} = -A$. Try making such a matrix for yourself. Now show that a skew-symmetric matrix is actually *not invertible* for odd $n$. 

(solution)

$$A = \begin{bmatrix}
 0 & -2 & -3 \\
 2 & 0 & 1 \\
 3 & -1 & 0
\end{bmatrix}$$

Here is an example of an odd $n$ skew-symmetric matrix. Is it invertible? How do we know?

Moving on:

Starting from the properties of the matrix:

$$A^T = -A$$

This is the claim

$$det(A^T) = det(-A)$$

On the left, $det(A^T) = det(A)$

and then using the rule regarding determinants of scalar multiples:

$$det(cA) = (c)^{n}det(A)$$

$$det(-1(A)) = (-1)^{n}det(A)$$

Holding both sides equal, we can see that the determinant can be nonzero for even $n$:

$$det(A) = (-1)^{2i}det(A) = det(A)$$ , but for odd $n$:

$$det(A) = (-1)^{2i+1}det(A) = -det(A)$$

The only determinant that can satisfy these constraints is zero, thus noninvertible. What was the determinant of the above matrix?

### 6. 

Determine values of k such that the system:

$$4x_1+kx_2 = 12$$

$$kx_1+x_2= 3$$

Is consistent.

(solution)

Assume $k \neq 0$ then we can do $R_2 - \dfrac{k}{4}R_1 \rightarrow R2$.


$$4x_1+kx_2 = 12$$

$$     (1-\dfrac{k^2}{4})x_2 = 3(1-k)$$

Therefore it must be true that:

$$x_2 = \dfrac{3(1-k)}{(1-\dfrac{k^2}{4})}$$

Meaning that $k \in \mathbb{R}, k \neq \{\pm2\}$. If we also check if $k = 0$, the system is still consistent.
