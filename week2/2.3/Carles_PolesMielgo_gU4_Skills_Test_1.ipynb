### 5. 

A $n \times n$ *square* matrix is **skew-symmetric** if $A^{T} = -A$. Try making such a matrix for yourself. Now show that a skew-symmetric matrix is actually *not invertible* for odd $n$. 

For n = 2, as an example:

$A$ = $\begin{bmatrix}
a_{11} & a_{12} \\
a_{21} & a_{22}\end{bmatrix}$

$A^{T}$ = $\begin{bmatrix}
a_{11} & a_{21} \\
a_{12} & a_{22}\end{bmatrix}$

$-A$ = $\begin{bmatrix}
-a_{11} & -a_{12} \\
-a_{21} & -a_{22}\end{bmatrix}$

If $-A$ = $A^{T}$:

$\begin{bmatrix}
-a_{11} & -a_{12} \\
-a_{21} & -a_{22}\end{bmatrix}$ = $\begin{bmatrix}
a_{11} & a_{21} \\
a_{12} & a_{22}\end{bmatrix}$

therefore, $-a_{12} = a_{21}$ and since $-a_{11}$ must be equal to $a_{11}$, this can only be true if $a_{11}$ = 0. Using the same reasoning, $a_{22}$ = 0.

So our matrix looks like:

$A$ = $\begin{bmatrix}
0 & a_{12} \\
-a_{12} & 0\end{bmatrix}$

In general:

$A = \begin{bmatrix}
    0 & a_{12} & a_{13} & \dots  & a_{1n} \\
    -a_{12} & 0 & a_{23} & \dots  & a_{2n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    -a_{1n} & -a_{2n} & -a_{3n} & \dots  & 0
\end{bmatrix}$

Now, showing that a skew-symmetric matrix is actually *not invertible* for odd $n$. 

If n = 2, for:

$A$ = $\begin{bmatrix}
-c & 0 \\
0 & -c\end{bmatrix}$

$det(A) = c^n * det(A)$

If n = 3, for:

$A$ = $\begin{bmatrix}
-c & 0 & 0\\
0 & -c & 0\\
0 & 0 & -c\end{bmatrix}$

$det(A) = -c^n * det(A)$

So, if c = 1, in general $det(A) = (-1)^n * det(A)$

If $A^{T} = -A$, then $det(A^{T}) = det(-A)$

$det(A) = (-1)^n * det(A)$

If n is even, $det(A) = (1) * det(A) = det(A)$

Per determinant properties we know that $det(A^{T}) = det(A)$, so if n is even, $det(A)$ exists, and therefore $det(A^{T}) = det(A)$

If n is odd $det(A) = (-1) * det(A) = det(A) = -det(A)$, and this is only true if det(A) = 0.

We also know that is a matrix has a determinant of 0, it cannot be inverted, so if n is odd, A cannot be inverted.

### 6. 

Determine values of k such that the system:

$$4x_1+kx_2 = 12$$

$$kx_1+x_2= 3$$

Is consistent.

Augmented matrix:

$\begin{bmatrix}
4 & k & 12 \\
k & 1 & 3 \end{bmatrix}$

Consistent means that the system must have solution.

1) $L2 = L2 - L1 * k/4$

$\begin{bmatrix}
4 & k & 12 \\
0 & (4-k^2)/4 & 3-3k \end{bmatrix}$

To be consistent, $(4-k^2)/4 = 0$ and $3-3k = 0$ OR $(4-k^2)/4 \neq 0$

$(4-k^2)/4 = 0$, $k = 2$ (taking both roots, 2 and -2).

If k = 2, the matrix is:

$\begin{bmatrix}
4 & 2 & 12 \\
2 & 1 & 3 \end{bmatrix}$

$L1 = L1/2$

$\begin{bmatrix}
2 & 1 & 6 \\
2 & 1 & 3 \end{bmatrix}$

$L2 = L2 - L1$

$\begin{bmatrix}
2 & 1 & 6 \\
0 & 0 & -3 \end{bmatrix}$

which means that $0x_2 = -3$, so the $x_2$ has no solution and the system is NOT consistent.

If k = -2, the matrix is:

$\begin{bmatrix}
4 & -2 & 12 \\
-2 & 1 & 3 \end{bmatrix}$

$L1 = L1/2$

$\begin{bmatrix}
2 & -1 & 6 \\
-2 & 1 & 3 \end{bmatrix}$

$L2 = L2 + L1$

$\begin{bmatrix}
2 & 1 & 6 \\
0 & 0 & 9 \end{bmatrix}$

which means that $0x_2 = 9$, so the $x_2$ has no solution and the system is NOT consistent.

So k cannot be 2 or -2 for the system to be consistent.
