## LU Decomposition Lab
Do the following by hand.

We will follow the general structure we outlined in the lecture

Given $A{\bf x} = b$ and $A = LU$:

0. Check for decomposability, pivot if necessary
1. Decompose A into L and U.
2. Let $L{\bf y} = b$. 
3. Solve for ${\bf y}$.
5. Recall ${\bf y} = U{\bf x}$.
6. Solve this equation to get ${\bf x}$.

## Problem
$$\begin{bmatrix} 1 & 6 & 2 \\ 2 & 12 & 5 \\ -1 & -3 & -1 \end{bmatrix}
\begin{bmatrix} x_{0} \\ x_{1} \\ x_{2} \end{bmatrix} = \begin{bmatrix} 9 \\ -4 \\ 17 \end{bmatrix}$$

### LU Decomposition
#### Check for decomposability (Step 1)
Write out the submatrices for $A$. You should have an $A_{1}$, $A_{2}$, and $A_{3}$.

$$A_{1} = 1, A_{2} = \begin{bmatrix} 1 & 6 \\ 2 & 12 \end{bmatrix}, A_{3} = \begin{bmatrix} 1 & 6 & 2 \\ 2 & 12 & 5 \\ -1 & -3 & -1 \end{bmatrix}$$

What are the determinants for each submatrix? 

$\lvert A_{1} \rvert = 1, \lvert A_{2} \rvert = 0, \lvert A_{3} \rvert = -3$.  

Can $A$ be decomposed in the current state?

No it cannot.

#### Pivot (Step 1)
Swap rows in $A$. Check the result for decomposability again. Once you've found a swap that works, move on to the next step.

One possible swap (make sure you swap b rows around as well):

$$\begin{bmatrix} 2 & 12 & 5 \\ -1 & -3 & -1 \\ 1 & 6 & 2  \end{bmatrix}
\begin{bmatrix} x_{0} \\ x_{1} \\ x_{2} \end{bmatrix} = \begin{bmatrix} -4 \\ 17 \\ 9 \end{bmatrix}$$

$\lvert A_{1} \rvert = 2, \lvert A_{2} \rvert = 6, \lvert A_{3} \rvert = -3$.  

#### Form (Step 2)
Write out the L and U matrices using variables ($A_{ij}$, $L_{ij}$, $U_{ij}$) and 1s/0s where necessary.

$$ \begin{bmatrix} A_{11} & A_{12} & A_{13} \\ A_{21} & A_{22} & A_{23} \\ A_{31} & A_{32} & A_{33} \end{bmatrix} = 
\begin{bmatrix} 1 & 0 & 0 \\ L_{21} & 1 & 0 \\ L_{31} & L_{32} & 1 \end{bmatrix}
\begin{bmatrix} U_{11} & U_{12} & U_{13} \\ 0 & U_{22} & U_{23} \\ 0 & 0 & U_{33} \end{bmatrix}$$

#### Equations (Step 2)
Write out $A = LU$ in a component by component basis. You should end up with a series of equations based on matrix multiplication.  
Ex. $A_{11} = 1 * U_{11}$

First row  
$A_{11} = 1 * U_{11}$  
$A_{12} = 1 * U_{12}$  
$A_{13} = 1 * U_{13}$  

Second Row  
$A_{21} = L_{21} * U_{11}$  
$A_{22} = L_{21} * U_{12} + 1 * U_{22}$  
$A_{23} = L_{21} * U_{13} + 1 * U_{23}$  

Third Row  
$A_{31} = L_{31} * U_{11}$  
$A_{32} = L_{31} * U_{12} + L_{32} * U_{22}$  
$A_{33} = L_{31} * U_{13} + L_{32} * U_{23} + 1 * U_{33}$  


#### Solve the equations above for all variables (Step 2)
Solve for each varaible. What are the final L and U matrices?

First row  
$2 = 1 * U_{11} \qquad \Rightarrow \qquad U_{11} = 2$  
$12 = 1 * U_{12} \qquad \Rightarrow \qquad U_{12} = 12$  
$5 = 1 * U_{13} \qquad \Rightarrow \qquad U_{13} = 5$  

Second Row  
$-1 = L_{21} * U_{11}$  
$-1 = L_{21} * 2 \qquad \Rightarrow \qquad L_{21} = -\frac{1}{2}$  
$-3 = L_{21} * U_{12} + 1 * U_{22}$  
$-3 = -\frac{1}{2} * 12 + 1 * U_{22} \qquad \Rightarrow \qquad U_{22} = 3$  
$-1 = L_{21} * U_{13} + 1 * U_{23}$  
$-1 = -\frac{1}{2} * 5 + 1 * U_{23} \qquad \Rightarrow \qquad U_{23} = \frac{3}{2}$  

Third Row  
$1 = L_{31} * U_{11}$  
$1 = L_{31} * 2 \qquad \Rightarrow \qquad L_{31} = \frac{1}{2}$  
$6 = L_{31} * U_{12} + L_{32} * U_{22}$  
$6 = \frac{1}{2} * 12 + L_{32} * 3 \qquad \Rightarrow \qquad L_{32} = 0$  
$2 = L_{31} * U_{13} + L_{32} * U_{23} + 1 * U_{33}$  
$2 = \frac{1}{2} * 5 + 0 * \frac{3}{2} + U_{33} \qquad \Rightarrow \qquad U_{33} = -\frac{1}{2}$ 

FINAL RESULTS:  
$$L = \begin{bmatrix} 1 & 0 & 0 \\ -\frac{1}{2} & 1 & 0 \\ \frac{1}{2} & 0 & 1 \end{bmatrix} \ 
U = \begin{bmatrix} 2 & 12 & 5 \\ 0 & 3 & \frac{3}{2} \\ 0 & 0 & -\frac{1}{2} \end{bmatrix}$$

Note: If you used a different pivot, your matrices will be different.

### Solving the system
#### Solve for ${\bf y}$ (Step 3/4)
Next, we set $L{\bf y} = b$.  
Plug in your $L$ matrix and $b$.  
You should be able to solve this system of equations for ${\bf y}$.

First, we set $L{\bf y} = b$, making sure to pivot $b$ the same way we pivoted $A$.

$$\begin{bmatrix} 1 & 0 & 0 \\ -\frac{1}{2} & 1 & 0 \\ \frac{1}{2} & 0 & 1 \end{bmatrix}
\begin{bmatrix} y_{1} \\ y_{2} \\ y_{3} \end{bmatrix} =
\begin{bmatrix} -4 \\ 17 \\ 9 \end{bmatrix}$$

Solve for ${\bf y}$:

$-4 = 1 * y_{1}$  
$-4 = y_{1}$

$17 = -\frac{1}{2} * y_{1} + 1 * y_{2}$  
$15 = y_{2}$

$9 = \frac{1}{2} * y_{1} + 0 * y_{2} + 1 * y_{3}$  
$11 = y_{3}$

This leaves us with:  
$${\bf y} = \begin{bmatrix} -4 \\ 15 \\ 11 \end{bmatrix}$$

#### Solve for ${\bf x}$ (Step 5/6)
Now that we have ${\bf y}$, recall that $U{\bf x} = {\bf y}$.  
Plug in your $U$ and ${\bf y}$ from above.  
You should be able to solve for ${\bf x}$.

First, we set $U{\bf x} = {\bf y}$.
$$\begin{bmatrix} 2 & 12 & 5 \\ 0 & 3 & \frac{3}{2} \\ 0 & 0 & -\frac{1}{2} \end{bmatrix}
\begin{bmatrix} x_{1} \\ x_{2} \\ x_{3} \end{bmatrix} = \begin{bmatrix} -4 \\ 15 \\ 11 \end{bmatrix}$$

Solve for ${\bf x}$ working backwards:

$11 = -\frac{1}{2} * x_{3}$  
$-22 = x_{3}$

$15 = 3 * x_{2} + \frac{3}{2} * x_{3}$  
$16 = x_{2}$

$-4 = 2 * x_{1} + 12 * x_{2} + 5 * x_{3}$  
$-43 = x_{1}$

Final solution:  
$${\bf x} = \begin{bmatrix} -43 \\ 16 \\ -22 \end{bmatrix}$$
